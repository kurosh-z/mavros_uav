```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def generate_circle_traj(radius=3.0,  numSteps = 60
                         , altitude=2):
   
    numPoints = 1 * numSteps
    # Create [x,y] points along a circle with specified radius
    PI = np.pi
    xpoints = np.flip(radius * np.sin(np.linspace(-PI,PI,num=numPoints)))
    ypoints =np.flip(radius * np.cos(np.linspace(-PI,PI,num=numPoints))-radius) 
    zpoints =np.flip(altitude*np.ones((numPoints, )))
 
    posi = list(zip(xpoints.tolist(), ypoints.tolist(), zpoints.tolist()))
    return [(0,0, 0)]+posi+[(0,0, 0)]
```

```python
posi = generate_circle_traj()
```

```python
fig , ax = plt.subplots()
ax.plot(np.array(posi)[50:-1,0], np.array(posi)[50:-1,1 ])
```
